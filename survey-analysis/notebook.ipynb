```python
import pandas as pd
from pathlib import Path
```

```python
en_survey_path = Path.cwd() / "data" / "form_responses_en.csv"
de_survey_path = Path.cwd() / "data" / "form_responses_de.csv"
out_survey_path = Path.cwd() / "data" / "form_responses_merged.csv"
```

```python
df_en = pd.read_csv(en_survey_path)
df_de = pd.read_csv(de_survey_path)
```

```python
# Generate column mapping de -> en
mapping = {}
for c_de, c_en in zip(df_de.columns, df_en.columns):
    mapping[c_de] = c_en
mapping
```

```python
# rename columns using mapping
df_de_renamed = df_de.rename(columns=mapping)
df_de_renamed.head()
```
